```python
import pandas as pd
from IPython.display import display, Markdown
```

```python

lifeExpectancyOriginal = pd.read_csv("/content/life_expectancy.csv")

fertilityRateOriginal = pd.read_csv("/content/fertility_rate.csv")

countryPopulationOriginal = pd.read_csv("/content/country_population.csv")

metadataCountryOriginal = pd.read_csv("/content/Metadata_Country.csv")
```

```python
lifeExpectancy = lifeExpectancyOriginal.copy()

fertilityRate = fertilityRateOriginal.copy()

countryPopulation = countryPopulationOriginal.copy()

metadataCountry = metadataCountryOriginal.copy()
```

```python
# Function to display DataFrame info with Markdown format

def formattedOutput(dataFrame, heading):
  display(Markdown(f"### {heading}"))
  display(dataFrame.head())
#  display(dataFrame.info())


formattedOutput(lifeExpectancy, "Life Expectancy Data")

formattedOutput(fertilityRate, "Fertility Rate Data")

formattedOutput(countryPopulation, "Country Population Data")

formattedOutput(metadataCountry, "MetaData Country")
```

```python
unnecessaryColumns = ["Indicator Name", "Indicator Code"]

dataFrames = {
    "lifeExpectancy": lifeExpectancy,
    "fertilityRate": fertilityRate,
    "countryPopulation": countryPopulation,
}

# Drop Columns
try:
  for Name, DataFrame in dataFrames.items():
    dataFrames[Name] = DataFrame.drop(columns=unnecessaryColumns, inplace=True)
except KeyError:
  pass
```

```python
# LIFE EXPECTANCY & FERTILITY RATE

dataFrames = [lifeExpectancy, fertilityRate]

for dataFrame in dataFrames:

  # Fill missing values w "Mean"

  for column in dataFrame.columns[2:]:

    # Calculate mean of column
    columnMean = dataFrame[column].mean()

    dataFrame[column].fillna(columnMean, inplace=True)
```

```python
# COUNTRY POPULATION

for column in countryPopulation.columns[2:]:

  # Calculate "Median"
  columnMedian = countryPopulation[column].median()

  countryPopulation[column].fillna(columnMedian, inplace=True)
```

**Data Type Conversion**

```python
# LIFE EXPECTANCY & FERTILITY RATE

for dataFrame in dataFrames:

  # Country Name, Country Code - String, title
  for column in dataFrame.columns[:3]:
    dataFrame[column] = dataFrame[column].astype(str).str.upper()

  # Floats upto 2ff
  for column in dataFrame.columns[2:]:
    dataFrame[column] = dataFrame[column].astype(float).round(2)
```

```python
# COUNTRY POPULATION

# Country Name, Country Code - String, title
for column in countryPopulation.columns[:3]:
  countryPopulation[column] = countryPopulation[column].astype(str).str.upper()

for column in countryPopulation.columns[2:]:
  countryPopulation[column] = countryPopulation[column].astype(float)
  countryPopulation[column] = countryPopulation[column].astype(int)
```

**Cleaning Metadata Country**

```python
metadataCountry = metadataCountry[["Country Code", "Region"]]
metadataCountry.head()
```

```python
# Melting - LIFE EXPECTANCY

years = [str(i) for i in range(1961, 2017, 1)]

lifeExpectancyMelted = pd.melt(lifeExpectancy, id_vars="Country Code", var_name="Year", value_vars=years, value_name="Life Expectancy")

# Melting - FERTILITY RATE LOL

fertilityRateMelted = pd.melt(fertilityRate, id_vars="Country Code", var_name="Year", value_vars=years, value_name="Fertility Rate")

# Melting Country Population

countryPopulationMelted = pd.melt(countryPopulation, id_vars="Country Code", var_name="Year", value_vars=years, value_name="Population")
```

**Merging**

```python
# Merge - Metadata Country & Country Population

mergedDataFrame = pd.merge(metadataCountry, countryPopulationMelted, how="left", on="Country Code")

# Merge - mergedDataFrame & Life Expectancy

mergedDataFrame = pd.merge(mergedDataFrame, lifeExpectancyMelted, how="left", on=["Country Code", "Year"])

# Merge - mergedDataFrame & Fertility Rate

mergedDataFrame = pd.merge(mergedDataFrame, fertilityRateMelted, how="left", on=["Country Code", "Year"])

mergedDataFrame.dropna(axis=0, inplace=True)
```
